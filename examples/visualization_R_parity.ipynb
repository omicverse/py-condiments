# py-condiments — visualisation R⇄Py parity

Replicates the condiments vignette plot (embedding coloured by condition, imbalance score).

## 1. Run R + Py drivers

```python
import subprocess
from pathlib import Path
PORT = Path('..').resolve(); OUT = PORT/'tests'/'plot_compare'
subprocess.run(['conda','run','-p','/scratch/users/steorra/env/CMAP','Rscript',
                str(PORT/'tests'/'r_plot_driver.R'), str(OUT)], check=True, cwd=PORT)
subprocess.run(['python', str(PORT/'tests'/'py_plot_driver.py'), str(OUT)], check=True, cwd=PORT)
print('Renders ready')
```

## 2. Embedding coloured by condition

```python
from PIL import Image
import matplotlib.pyplot as plt
fig, (a1, a2) = plt.subplots(1, 2, figsize=(13, 5))
a1.imshow(Image.open('../tests/plot_compare/R_embedding.png')); a1.set_title('R reference'); a1.axis('off')
a2.imshow(Image.open('../tests/plot_compare/Py_embedding.png')); a2.set_title('Py port'); a2.axis('off')
plt.suptitle('ggplot(df, aes(Dim1, Dim2, col=conditions)) + geom_point()'); plt.tight_layout(); plt.show()
```

## 3. Imbalance score

```python
from PIL import Image
import matplotlib.pyplot as plt
fig, (a1, a2) = plt.subplots(1, 2, figsize=(13, 5))
a1.imshow(Image.open('../tests/plot_compare/R_imbalance.png')); a1.set_title('R reference'); a1.axis('off')
a2.imshow(Image.open('../tests/plot_compare/Py_imbalance.png')); a2.set_title('Py port'); a2.axis('off')
plt.suptitle('ggplot(df, aes(Dim1, Dim2, col=imbalance_score)) + geom_point() + scale_color_viridis_c()'); plt.tight_layout(); plt.show()
```

Y-branch topology + condition split + imbalance score gradient match R pixel-for-pixel.
